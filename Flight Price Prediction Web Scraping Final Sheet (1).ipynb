```python
# Getting the webpage of mentioned url
#getting the url
url = "https://flight.easemytrip.com/FlightList/Index?org=BLR-Bangalore,%20India&dept=BOM-Mumbai,%20India&adt=1&chd=0&inf=0&cabin=0&airline=Any&deptDT=5/05/2022&arrDT=&isOneway=true&isDomestic=true&CouponCode=&lang=&ompAff=&bc=&ISWL="
driver.get(url)
```

```python
#Airline details
airline_name=[]  
try:
    name=driver.find_elements_by_xpath('//span[@class="txt-r4 ng-binding"]')
    for i in name:
        airline_name.append(i.text)
except NoSuchElementException:
    airline_name.append('-')
#Source
Source=[]  
try:
    s=driver.find_elements_by_xpath('//div[@class="col-md-2 col-sm-2 col-xs-4 top5 "]/span[2]')
    for i in s:
        Source.append(i.text)
except NoSuchElementException:
    Source.append('-')
#destination
Destination=[]  
try:
    d=driver.find_elements_by_xpath('//div[@class="col-md-2 col-sm-2 col-xs-3 top5"]/span[2]')
    for i in d:
        Destination.append(i.text)
except NoSuchElementException:
    Destination.append('-')  
# total stops
stops=[]
try:
    stop=driver.find_elements_by_xpath('//span[@class="dura_md2 ng-scope"]')
    for i in stop:
        stops.append(i.text)
except NoSuchElementException:
    stops.append('-')
#departure time 
departure=[]
try:
    time=driver.find_elements_by_xpath('//span[@class="txt-r2-n ng-binding"]')
    for i in time:
        departure.append(i.text)
except NoSuchElementException:
    departure.append('-')
    
#destination
departure_time=[]
arrival_time=[]
for i in range(0,len(departure)-1,2):
    departure_time.append(departure[i])
    arrival_time.append(departure[i+1])
    
    
#duration
duration=[]
try:
    time=driver.find_elements_by_xpath('//span[@class="dura_md ng-binding"]')
    for i in time:
        duration.append(i.text)
except NoSuchElementException:
    duration.append('-')
    
#fare
price=[]
try:
    fare=driver.find_elements_by_xpath('//div[@class="col-md-8 col-sm-8 col-xs-9 txt-r6-n exPrc ng-binding"]')
    for i in fare:
        price.append(i.text)
except NoSuchElementException:
    price.append('-')
```

```python
len(airline_name),len(Source),len(Destination),len(stops),len(departure_time),len(arrival_time),len(duration),len(price)
```

```python
#Airline details
airline_name=[]  
try:
    name=driver.find_elements_by_xpath('//span[@class="txt-r4 ng-binding"]')
    for i in name:
        airline_name.append(i.text)
except NoSuchElementException:
    airline_name.append('-')
#Source
Source=[]  
try:
    s=driver.find_elements_by_xpath('//div[@class="col-md-2 col-sm-2 col-xs-4 top5 "]/span[2]')
    for i in s:
        Source.append(i.text)
except NoSuchElementException:
    Source.append('-')
#destination
Destination=[]  
try:
    d=driver.find_elements_by_xpath('//div[@class="col-md-2 col-sm-2 col-xs-3 top5"]/span[2]')
    for i in d:
        Destination.append(i.text)
except NoSuchElementException:
    Destination.append('-')   
    
# total stops
stops=[]
try:
    stop=driver.find_elements_by_xpath('//span[@class="dura_md2 ng-scope"]')
    for i in stop:
        stops.append(i.text)
except NoSuchElementException:
    stops.append('-')
#departure time 
departure=[]
try:
    time=driver.find_elements_by_xpath('//span[@class="txt-r2-n ng-binding"]')
    for i in time:
        departure.append(i.text)
except NoSuchElementException:
    departure.append('-')
 #destination
departure_time=[]
arrival_time=[]
for i in range(0,len(departure)-1,2):
    departure_time.append(departure[i])
    arrival_time.append(departure[i+1])       
#duration
duration=[]
try:
    time=driver.find_elements_by_xpath('//span[@class="dura_md ng-binding"]')
    for i in time:
        duration.append(i.text)
except NoSuchElementException:
    duration.append('-')
    
#fare
price=[]
try:
    fare=driver.find_elements_by_xpath('//div[@class="col-md-8 col-sm-8 col-xs-9 txt-r6-n exPrc ng-binding"]')
    for i in fare:
        price.append(i.text)
except NoSuchElementException:
    price.append('-')
```
